## check for null values 

```python
df.isnull().values.any()   # isnull() checks all data for any null any() will return true if null is found 
```

```python
def plot_corr(df, size =11):
    """
        input:: 
          df- pandas framework 
          size: vertical and horizontal size of the plot 
          
        Output: 
            matix of correlation between columns. 
    """
    corr = df.corr() # data frame correlation function 
    fig, ax = plt.subplots (figsize =(size, size)) 
    ax.matshow(corr) # color code the rectangels by correlation value 
    plt.xticks(range(len(corr.columns)), corr.columns) # draw x tick marks 
    plt.yticks(range(len(corr.columns)), corr.columns) # draw y tick marks 
```

```python
plot_corr(df)
```

```python
df.corr() # 1.0000 means perfect correlation but skin and thickness had correlation and should be removed 
```
